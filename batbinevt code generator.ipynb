```python
#Importing the necessary libraries

import pandas as pd
```

```python
df = pd.read_excel('Yearly GRB data/2006.xlsx')       # The input file here would be one of the files from the yearly GRB data (2006,2007,....2020)
                                      # We will read the input file and store it as a Pandas dataframe.
```

```python
df.shape, df.shape[0]       #Find the shape of the dataframe
```

```python
#We will delete those columns which we don't need for this code

df.drop(["a"],axis=1,inplace=True)
df.drop(["b"],axis=1,inplace=True) 
df.drop(["c"],axis=1,inplace=True) 
df.drop(["d"],axis=1,inplace=True) 
df.drop(["e"],axis=1,inplace=True) 
df.drop(["f"],axis=1,inplace=True) 
df.drop(["g"],axis=1,inplace=True) 
```

```python
df     # The new dataframe contains only the trigger number, lower energy value (e1) and upper energy value (e2)
```

```python
E_a=df["e1"]
E_b=df["e2"]
```

```python
# Now we generate the respective BATBINEVT commands for each of the GRB for a year's data.


E1=[]
E2=[]
num=[]



for i in range(0,df.shape[0]):
    num.append(df["trigger"][i])
    E1.append(E_a[i])
    E2.append(E_b[i])

    print("\n batbinevt infile='./"+ "00"+str(num[i])+ "000"+"-results/events/sw"+ "00" +str(num[i]) + "000"+"b_all.evt' outfile='./" + "00"+ str(num[i]) + "000"+ "-results/lc/test.lc' outtype=LC timedel=0.016 timebinalg=u energybins='" +str(E1[i])+ "-"+ str(E2[i])+ "' detmask='./" + "00"+ str(num[i])+ "000"+ "-results/auxil/sw" + "00"+ str(num[i])+ "000"+ "b_qmap.fits' ecol=ENERGY weighted=YES outunits=RATE clobber=yes")

```
