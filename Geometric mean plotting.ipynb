```python
# This program takes in the values of geometric means and simply plots them

import numpy as np
import pandas as pd
from astropy.io import ascii
from scipy.optimize import curve_fit
from matplotlib.pylab import plt

data=pd.read_excel('Geometric mean data.xlsx')       # importing the already calculated GM data (calculated by the Geometric mean.ipynb file)

```

```python
data
```

```python
#Plotting weighted means for T90
#Some Matplotlib settings
from pylab import rcParams
rcParams['figure.figsize'] = 30, 20
plt.rc('axes', labelsize='45')
plt.xticks(fontsize='40')
plt.yticks(fontsize='40')

z_GM=data['z_GM']
T90_GM=data['T90_GM']
T90_GM_error=data['T90_GM_error']
z_err=data['z_err']


n,bin_edges=np.histogram(z_GM,bins=6)
xbin=0.5*(bin_edges[1]-bin_edges[0])
plt.errorbar(z_GM,T90_GM,yerr=T90_GM_error,xerr=xbin,fmt='.k',elinewidth=4,capsize=6,capthick=4)
plt.scatter(z_GM, T90_GM, s = 300)

plt.yscale('log')
plt.xscale('log')
plt.xlabel('z',fontweight='bold')
plt.ylabel('T90',fontweight='bold')
plt.savefig('geometric_mean_grb90.pdf')
```
